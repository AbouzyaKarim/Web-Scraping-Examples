```python
import requests
from bs4 import BeautifulSoup
```

```python
result = requests.get("https://www.google.com/")
print(result.status_code)
```

print headers of request result

```python
print(result.headers)
```

print content of the web page

```python
src=result.content
print(src)
```

parse and process the source with beautifulsoup

```python
soup = BeautifulSoup(src,'lxml')
```

we find and find_all methods to acess specific information

```python
links = soup.find_all("a")
print(links)
print("-----------")
```

```python
for l in links:
    #print(l)
    if "Images" in l.text:
        print(l)
        print(l.attrs['href'])
```
